```python
import pandas as pd
```

```python
data = pd.read_excel("Country_and_Territory_Ratings_and_Statuses_FIW_1973-2024 (1).xlsx", sheet_name='Country Ratings, Statuses ', header=1)
```

```python
data_territories = pd.read_excel("Country_and_Territory_Ratings_and_Statuses_FIW_1973-2024 (1).xlsx", sheet_name='Territory Ratings, Statuses', header=1)
```

```python
def rename_columns(df):
    # drop the first row
    df = df.iloc[1: , :]
    # name unnamed columns
    # https://www.geeksforgeeks.org/how-to-get-column-names-in-pandas-dataframe/
    column_list = list(df.columns)
    for index, name in enumerate(column_list):
        if index == 0:
            continue
        if index % 3 == 1:
            old_col = name
            if type(name) == str:
                new_col = f"{name[4:8]}_PR"
            else:
                new_col = f"{old_col}_PR"
            new_col = f"{old_col}_PR"
            df = df.rename(columns={old_col: new_col})
        if index % 3 == 2:
            old_col = name
            new_col = f"{df.columns[(index - 1)][:-2]}CL"
            df = df.rename(columns={old_col: new_col})
        if index % 3 == 0:
            old_col = name
            new_col = f"{df.columns[(index - 2)][:-2]}Status"
            df = df.rename(columns={old_col: new_col})
    
    return df
    
```

```python
# clean the dataframes
data = rename_columns(data)
data_territories = rename_columns(data_territories)

# combine countries and territories
data_full = pd.concat([data, data_territories])

# rename countries to match the geojson
data_full.loc[data_full["Year(s) Under Review"] == "United States", "Year(s) Under Review"] = "United States of America"
data_full.loc[data_full["Year(s) Under Review"] == "Serbia", "Year(s) Under Review"] = "Republic of Serbia"
data_full.loc[data_full["Year(s) Under Review"] == "North Macedonia", "Year(s) Under Review"] = "Macedonia"
data_full.loc[data_full["Year(s) Under Review"] == "Eswatini", "Year(s) Under Review"] = "Swaziland"
data_full.loc[data_full["Year(s) Under Review"] == "The Gambia", "Year(s) Under Review"] = "Gambia"
data_full.loc[data_full["Year(s) Under Review"] == "Guinea-Bissau", "Year(s) Under Review"] = "Guinea Bissau"
data_full.loc[data_full["Year(s) Under Review"] == "Cote d'Ivoire", "Year(s) Under Review"] = "Ivory Coast"
data_full.loc[data_full["Year(s) Under Review"] == "Tanzania", "Year(s) Under Review"] = "United Republic of Tanzania"
data_full.loc[data_full["Year(s) Under Review"] == "Congo (Brazzaville)", "Year(s) Under Review"] = "Republic of Congo"
data_full.loc[data_full["Year(s) Under Review"] == "Congo (Kinshasa)", "Year(s) Under Review"] = "Democratic Republic of the Congo"

new_row = data_full.loc[data_full["Year(s) Under Review"] == "Denmark"]
new_row.loc[data_full["Year(s) Under Review"] == "Denmark", "Year(s) Under Review"] = "Greenland"
data_full = pd.concat([data_full, new_row], ignore_index=True)

for columns in data_full.columns[16:]:
    print(columns)

# data_full.to_csv('freedom_scores.csv')
```
